```python
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='MDVP:Fo(Hz)', hue='status', kde=True)
plt.title('Distribution of Average Vocal Fundamental Frequency')
plt.show()
```


* The average vocal fundamental frequency represents the average rate of vocal cord vibrations during speech. Changes in this frequency could be indicative of certain voice disorders, including Parkinson's Disease.

* A significant shift in the central tendency (mean or median) of this frequency between the two groups might suggest a distinguishable marker for Parkinson's.

* Overlap or Separation: Observing whether there's an overlap or separation in the distributions of this frequency for healthy and Parkinson's individuals could indicate the potential usefulness of this feature in classifying the two groups.


* The distribution of average vocal fundamental frequency for healthy patients (status 0) is concentrated between 150 Hz and 200 Hz, with a mean of 170 Hz.
* The distribution of average vocal fundamental frequency for patients with Parkinson's disease (status 1) is broader and shifted to the left, with a mean of 135 Hz.
* This suggests that patients with Parkinson's disease tend to have lower average vocal fundamental frequency than healthy patients.
* The overlap between the two distributions also suggests that there is some variability in average vocal fundamental frequency among both healthy patients and patients with Parkinson's disease.
* The graph shows that there is no single cutoff value for average vocal fundamental frequency that can be used to definitively diagnose Parkinson's disease. However, a lower average vocal fundamental frequency may be a sign of the disease, especially in conjunction with other symptoms.
* The graph also shows that there is some overlap in the distribution of average vocal fundamental frequency between healthy patients and patients with Parkinson's disease. This means that it is possible for a patient to have a lower average vocal fundamental frequency and still be healthy.

```python
jitter_cols = ['MDVP:Jitter(%)', 'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP']
plt.figure(figsize=(12, 8))
for col in jitter_cols:
    sns.histplot(data=data, x=col, hue='status', kde=True, bins=20, alpha=0.5)
plt.legend(title='Status', loc='upper right', labels=['Healthy', 'Parkinsons'])
plt.title('Distribution of Jitter Measures')
plt.show()
```



* Jitter refers to the cycle-to-cycle variation in the duration of speech sounds. In voice analysis, jitter measures variations in the fundamental frequency, which can indicate vocal fold irregularities.

* Understanding how these measures distribute across the two groups can unveil if there's a significant difference or if they exhibit separate distributions.

* Parkinson's Disease can affect vocal cord control, resulting in higher jitter measures due to irregular vocal fold movements. A graph displaying these measures might highlight increased variability in individuals with Parkinson's compared to healthy individuals.

* The graph shows that the distribution of jitter measures is significantly different between the two groups. Healthy patients have a lower mean jitter measure (0.015) than patients with Parkinson's disease (0.025). Additionally, the distribution of jitter measures is narrower for healthy patients than for patients with Parkinson's disease. This suggests that patients with Parkinson's disease have more variability in their jitter measures than healthy patients.

* The graph also shows that there is some overlap in the distribution of jitter measures between the two groups. This means that it is possible for a patient to have a high jitter measure and still be healthy. However, a high jitter measure is more likely to be seen in patients with Parkinson's disease.

* The graph suggests that jitter measures can be used to distinguish between healthy patients and patients with Parkinson's disease with a high degree of accuracy.

* The graph also suggests that jitter measures may be useful for tracking the progression of Parkinson's disease and assessing the effectiveness of treatment.

```python
shimmer_cols = ['MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5', 'MDVP:APQ', 'Shimmer:DDA']

plt.figure(figsize=(16, 10))

# Reshape the data for boxplot visualization
data_melted_shimmer = pd.melt(data, id_vars='status', value_vars=shimmer_cols, var_name='Attribute', value_name='Value')

# Define custom colors for the status categories
colors = ["#1f77b4", "#ff7f0e"]  # Blue for Healthy, Orange for Parkinsons

sns.boxplot(data=data_melted_shimmer, x='Attribute', y='Value', hue='status', palette=colors)
plt.title('Boxplot of Amplitude Variation Measures')
plt.xlabel('Amplitude Variation Attributes')
plt.ylabel('Value')

plt.xticks(rotation=45)
plt.show()
```

Visualizing boxplots for amplitude variation measures provides valuable insights into vocal irregularities associated with Parkinson's Disease. It aids in comparing these measures between healthy individuals and those affected, identifying potential outliers, and selecting important features for disease prediction or classification models.

* MDVP:Shimmer: This measurement quantifies the amount of variation in the amplitude of the vocal signal. Parkinson's disease patients tend to have higher MDVP:Shimmer values than healthy patients, which suggests that their voices are more breathy and unstable.

* MDVP:Shimmer(dB): This measurement is a log-transformed version of MDVP:Shimmer, which makes it easier to compare values across different subjects. Parkinson's disease patients also tend to have higher MDVP:Shimmer(dB) values than healthy patients.

* Shimmer:APQ3: This measurement quantifies the percentage of vocal cycles in a given speech sample that have a shimmer value greater than the third quartile of all shimmer values. Parkinson's disease patients tend to have higher Shimmer:APQ3 values than healthy patients, which suggests that their voices have more high-amplitude shimmer.

* Shimmer:APQ5: This measurement is similar to Shimmer:APQ3, but it uses the fifth quartile instead of the third quartile. Parkinson's disease patients also tend to have higher Shimmer:APQ5 values than healthy patients.

* MDVP:APQ: This measurement quantifies the percentage of vocal cycles in a given speech sample that have a pitch value greater than the third quartile of all pitch values. Parkinson's disease patients tend to have lower 

* MDVP:APQ values than healthy patients, which suggests that their voices have fewer high-frequency components.

* Shimmer:DDA: This measurement quantifies the amount of jitter (variation in vocal cycle duration) in the voice signal. Parkinson's disease patients tend to have higher Shimmer:DDA values than healthy patients, which suggests that their voices are more jittery and unstable.

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python

X = data.drop(['name', 'status'], axis=1) #x features & y target
y = data['status']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression #binary
from sklearn.svm import SVC  #support vector classifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
logreg_pred = logreg.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, logreg_pred))
print("Logistic Regression Report:")
print(classification_report(y_test, logreg_pred))
```

Logistic Regression is a popular choice for binary classification problems, like predicting whether someone has Parkinson's Disease or not (0 for healthy, 1 for Parkinson's).

```python
from sklearn.metrics import accuracy_score
accuracy_logreg = accuracy_score(y_test, logreg_pred)
accuracy_logreg

```

```python

from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, labels):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

labels = ['Healthy', 'Parkinsons']


conf_matrix_logreg = confusion_matrix(y_test, logreg_pred)
print("Confusion Matrix - Logistic Regression:")

plot_confusion_matrix(conf_matrix_logreg, labels)
```

```python
svm = SVC()  # check test is positive or not
svm.fit(X_train, y_train)
svm_pred = svm.predict(X_test)
print("SVM Accuracy:", accuracy_score(y_test, svm_pred))
print("SVM Report:")
print(classification_report(y_test, svm_pred))
```

SVMs can capture non-linear relationships between features and the target variable, allowing them to model complex decision boundaries that might exist in the data. This is crucial as relationships in medical data can often be non-linear.
Medical datasets can often be imbalanced, with fewer instances of the target class (e.g., fewer instances of Parkinson's Disease). SVMs can handle imbalanced datasets well by focusing on the support vectors, which are the critical data points influencing the decision boundary.

```python
accuracy_svm = accuracy_score(y_test, svm_pred)
accuracy_svm
```

```python

def plot_confusion_matrix(cm, labels):
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

labels = ['Healthy', 'Parkinsons']
conf_matrix_svm = confusion_matrix(y_test, svm_pred)
print("\nConfusion Matrix - SVM:")

plot_confusion_matrix(conf_matrix_svm, labels)
```

```python
knn = KNeighborsClassifier() #knearestneighbours
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)
print("KNN Accuracy:", accuracy_score(y_test, knn_pred))
print("KNN Report:")
print(classification_report(y_test, knn_pred))
```

```python
accuracy_knn = accuracy_score(y_test, knn_pred) #highest accuracy score
accuracy_knn
```

```python
xgb = XGBClassifier() #xg boost (decision tree)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
print("XGBoost Report:")
print(classification_report(y_test, xgb_pred))
```

```python
accuracy_xgb = accuracy_score(y_test, xgb_pred)
accuracy_xgb
```
